```python
import numpy as np
from sklearn.linear_model import LinearRegression
training_data=np.loadtxt('0000000000002329_training_diabetes_x_y_train.csv', delimiter=',')
testing_data=np.loadtxt('0000000000002329_test_diabetes_x_test.csv', delimiter=',')
print(testing_data.shape)
print(training_data.shape)
```

```python
training_data
```

```python
testing_data
```

```python
X_test=testing_data
X_train=training_data[:, :10]
Y_train=training_data[:, 10:]
print(X_test.shape)
print(Y_train.shape)
print(X_train.shape)
```

```python
model=LinearRegression()
model.fit(X_train, Y_train)
ans=model.predict(X_test)
np.savetxt('predicted_data.csv', ans, delimiter=',')
```

```python
import matplotlib.pyplot as plt
plt.plot(X_train,Y_train)
plt.show()
```

```python
plt.plot(X_test,ans)
plt.show()
```

```python
feature_importance=np.abs(model.coef_[0])
import matplotlib.pyplot as plt
plt.bar(range(X_train.shape[1]),feature_importance)
plt.xlabel('Feature Index')
plt.ylabel('Feature Importance (Coefficient Magnitude)')
plt.title("Importance of Each Feature")
plt.show()
```
